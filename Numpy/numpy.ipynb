```python
'''
- Indexing means accessing specific elements from a NumPy array.
- Slicing means extracting a range or part of the array.
- Indexing starts from 0 (just like in Python lists).
- Works for 1D, 2D, and multi-dimensional arrays.
- Helps in data manipulation and selection. '''

# Indexing in 1D Array
arr = np.array([10, 20, 30, 40, 50])
print(arr[0]) # First element
print(arr[3]) # Fourth element
#Negative Indexing - Negative indexes count from the end. - 1 ‚Üí last element, 2 ‚Üí second last
print(arr[-1]) # Last element
print(arr[-3]) # Third element from the end

# Indexing in 2D Array
arr2d = np.array([[10, 20, 30],
[40, 50, 60]])
print(arr2d[0, 1]) # Row 0, Column 1 - 20
print(arr2d[0, 0]) # Row 0, Column 0 - 10
print(arr2d[1, 2]) # Row 1, Column 2 - 60
```

```python
# Slicing in 1D Array
''' 
Syntax: arr[start:end]
start is included, end is excluded.
'''
arr = np.array([10, 20, 30, 40, 50, 60])
print(arr[1:4]) # From index 1 to 3
print(arr[:3]) # From start to index 2
print(arr[3:]) # From index 3 to end

# Slicing in 2D Array
''' 
Syntax: arr[row_start:row_end, col_start:col_end]
Selects a sub-matrix from the array.
'''
arr2d = np.array([[10, 20, 30],[40, 50, 60],[70, 80, 90]])
print(arr2d.ndim)
print(arr2d[0:2, 1:3]) # Rows 0‚Äì1, Columns 1‚Äì2

# Step Value in Slicing
''' 
Syntax: arr[start:end:step]
Step defines how many elements to skip.
'''
arr = np.array([10, 20, 30, 40, 50, 60, 70])
print(arr[::2]) # Every 2nd element
print(arr[::-1]) # reverse 
print(arr[1:4]) # Every 2nd element
print(arr[1:6:2]) # From 1 to 5, step of 2

# Modifying Values using Indexing
''' 
You can update elements directly using indexes.
NumPy arrays are mutable (can be changed).
'''
arr = np.array([1, 2, 3, 4, 5])
arr[2] = 99
print(arr)
```

```python
''' 
Vectorization means performing operations on entire arrays instead of using
loops.
It allows you to write clean, faster, and more efficient code.
NumPy automatically applies operations element-wise on arrays.
'''
arr1 = np.array([1, 2, 3, 4])
arr2 = np.array([10, 20, 30, 40])
addition = arr1 + arr2
Subtraction = arr2 - arr1 
Multiplication =  arr1 * arr2
Division  = arr2 / arr1 
Power =  arr1 ** 2 
print(addition)
print(Subtraction)
print(Multiplication)
print(Division)
print(Power)

# Vectorization with Functions
arr = np.array([1, 2, 3, 4, 5])
print(np.sqrt(arr)) # Square root
print(np.square(arr)) # Square
print(np.exp(arr)) # Exponential
print(np.log(arr)) # Logarithm
''' 
These functions work on entire arrays at once.
No need for manual loops ‚Äî fully vectorized.
'''
```

```python
# Joining 1D Arrays using concatenate()
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])
result = np.concatenate((arr1, arr2))
print(result)
''' 
concatenate() joins arrays along an existing axis.
For 1D arrays, it joins them end to end.
'''
# Joining 2D Arrays using concatenate()
'''axis=0 ‚Üí join vertically (rows).
axis=1 ‚Üí join horizontally (columns).'''
arr1 = np.array([[1, 2],[3, 4]])
arr2 = np.array([[5, 6],[7, 8]])
result1 = np.concatenate((arr1, arr2), axis=0) # Row-wise
result2 = np.concatenate((arr1, arr2), axis=1) # Column-wise
print("Row-wise:\n", result1)
print("Column-wise:\n", result2)
```

```python
# Joining Arrays using hstack() and vstack()
''' hstack() joins arrays side by side (columns).
vstack() joins arrays top to bottom (rows).'''
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])
h = np.hstack((arr1, arr2))
v = np.vstack((arr1, arr2))
print("Horizontal Stack:", h)
print("Vertical Stack:\n", v)
```

```python
# Using np.stack() for New Dimensions
''' stack() joins arrays along a new axis.
Here, a new column dimension is created.'''
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])
stacked = np.stack((arr1, arr2), axis=1)
print(stacked)
```

```python
# Splitting 1D Array using split()
''' np.split(array, 3) splits array into 3 equal parts.
The number of elements must be divisible by parts.'''
arr = np.array([10, 20, 30, 40, 50, 60])
result = np.split(arr, 3)
print(result)
```

```python
# Unequal Splitting using array_split()
'''array_split() allows unequal splitting.
Useful when total elements can‚Äôt be divided equally.'''
arr = np.array([1, 2, 3, 4, 5])
result = np.array_split(arr, 3)
print(result)
```

```python
# Splitting 2D Arrays
'''hsplit() ‚Üí splits horizontally (columns).
vsplit() ‚Üí splits vertically (rows).'''
arr2d = np.array([[10, 20, 30, 40],
[50, 60, 70, 80]])
print(np.hsplit(arr2d, 2)) # Split columns
print(np.vsplit(arr2d, 2)) # Split rows
```

```python
'''Broadcasting allows NumPy to perform operations on arrays of different
shapes and sizes.
Normally, arrays must have the same shape to perform element-wise
operations.
Broadcasting automatically expands smaller arrays to match larger ones.
Makes operations faster, simpler, and memory-efficient.
Used heavily in machine learning and numerical computation.'''
# NumPy cannot add arrays with incompatible shapes.
# a.shape = (3,) and b.shape = (2,) ‚Üí ‚ùå Not compatible.
```

```python
'''
# Broadcasting Works When Shapes Match Rules
Broadcasting works when:

Broadcasting in NumPy 1

1. The arrays have same number of dimensions, or
2. The smaller array can expand along dimension = 1, or
3. The dimensions are compatible from right to left.
'''
# Scalar and Array
'''Scalar 10 is broadcasted to each element of the array.
Equivalent to [1,2,3,4] + [10,10,10,10] .'''
arr = np.array([1, 2, 3, 4])
result = arr + 10
print(result)

# 1D and 2D Arrays
a = np.array([[1, 2, 3],
[4, 5, 6]])
b = np.array([10, 20, 30])
result = a + b
print(result)
'''b (1D array) is broadcasted across each row of a .
NumPy repeats b to match the 2D shape of a .'''

# Column Broadcasting
'''Shapes Before Broadcasting: 
Array A: (2, 3)
Array B: (3,)'''
# NumPy expands B ‚Üí (1, 3) ‚Üí (2, 3)
''' Final Operation:
A + B
Each row of A adds with B.
üß† Rule:
Start from rightmost dimension.
Dimensions must be equal or one of them is 1.'''
a = np.array([[1, 2, 3],
[4, 5, 6]])
b = np.array([[10],
[20]])
result = a + b
print(result)
'''b is broadcasted column-wise.
NumPy expands b to match the shape of a .'''
```
